```python
import math
```

```python
#Класс фигура

class Shape :
    def __init__(self, x=0, y=0, z=0):
        self.x = x
        self.y = y
        self.z = z 
    

    #Методо для вычисления раастояни до другой фигуры

    def distance_to (self, other):
        dx = self.x - other.x 
        dy = self.y - other.y
        dz = self.z - other.z
        return math.sqrt(dx**2 + dy**2 + dz**2)
    

    #Метод для перемещения объекта

    def move(self, dx, dy, dz):
        self.x += dx
        self.y += dy
        self.z += dz

    
    def calculate_volume(self):
        return 0
    
    
    def __str__(self):
        return f"Shape at ({self.x}, {self.y}, {self.z})"
```

```python
# Для точки не надо конструктор и она будет использовать конструктор Shape

class Point(Shape):
    def __init__(self, x, y, z=0):
        self.x = x
        self.y = y
        self.z = z
    

    def calculate_volume(self):
        return 0
```

```python
# Класс отрезок

class Segment(Shape):
    def __init__(self, x, y, z, len):
        super().__init__(x, y, z) # будет вызывать наш конструктор
        self.len = len


    def calculate_volume(self):
        return 0
    

    def get__and__point(self):
        end_x = self.x + self.len
        return Point(end_x, self.y, self.z)
    

    def __str__(self):
        return f"Segment from ({self.x}, {self.y}, {self.z}) with len {self.len}"
```

```python
# класс прямоугольник

class Rectangle(Shape):
    def __init__(self, x, y, z, len, width):
         super().__init__(x, y, z)
         self.len = len
         self.width = width


    # Вычисляем площадь

    def calculate_volume(self):
        return self.len * self.width 
        
    # для вычисления диогонали

    def get_diagonal(self):
        return math.sqrt(self.len**2 + self.width**2)
    

    def __str__(self):
        return f"Rectangle at ({self.x}, {self.y}, {self.z}) with dimensions {self.len}x{self.width}"
```

```python
# Класс Параллелепипед

class Box(Shape):
    def __init__(self, x, y, z, len, width, height):
        super().__init__(x, y, z)
        self.len = len
        self.width = width
        self.height = height


    # Объем

    def calculate_volume(self):
        return self.len * self.width * self.height 
    

    def get_diagonal(self):
        return math.sqrt(self.len**2 + self.width**2 + self.height**2)
    

    def __str__(self):
        return f"Box at ({self.x}, {self.y}, {self.z}) with dimensions {self.len}x{self.width}x{self.height}"
       
```

```python
print("=" * 60)
print("ПРОВЕРКА РАБОТЫ ВСЕХ КЛАССОВ")
print("=" * 60)

# Создаем объекты (экземпляры классов)
print("1. Создаем объекты:")
p1 = Point(1, 2, 3)
print(f"Точка: {p1}")

seg1 = Segment(0, 0, 0, 5)
print(f"Отрезок: {seg1}")

rect1 = Rectangle(10, 10, 0, 4, 6)
print(f"Прямоугольник: {rect1}")

box1 = Box(0, 0, 0, 3, 4, 5)
print(f"Параллелепипед: {box1}")

print("\n2. Проверяем наследование:")
print(f"Точка является Shape: {isinstance(p1, Shape)}")
print(f"Отрезок является Shape: {isinstance(seg1, Shape)}")
print(f"Прямоугольник является Shape: {isinstance(rect1, Shape)}")
print(f"Параллелепипед является Shape: {isinstance(box1, Shape)}")

print("\n3. Вычисляем объем (полиморфизм в действии):")
print(f"Объем точки: {p1.calculate_volume()}")
print(f"Объем отрезка: {seg1.calculate_volume()}")  # Теперь работает!
print(f"Площадь прямоугольника: {rect1.calculate_volume()}")
print(f"Объем параллелепипеда: {box1.calculate_volume()}")

print("\n4. Используем другие методы:")
print(f"Диагональ прямоугольника: {rect1.get_diagonal():.2f}")
print(f"Диагональ параллелепипеда: {box1.get_diagonal():.2f}")

# Проверяем конечную точку отрезка
end_point = seg1.get__and__point()
print(f"Конечная точка отрезка: {end_point}")

# Проверяем расстояние между фигурами
print(hasattr(box1, 'distance_to'))

print("\n5. Перемещаем фигуры:")
print(f"До перемещения: {rect1}")
rect1.move(5, 5, 0)
print(f"После перемещения: {rect1}")

print("\n6. Математическая проверка правильности расчетов:")
# Проверяем диагональ прямоугольника 4x6: √(4² + 6²) = √(16 + 36) = √52 ≈ 7.21
calculated_diag = rect1.get_diagonal()
expected_diag = math.sqrt(4**2 + 6**2)
print(f"Диагональ прямоугольника: {calculated_diag:.2f}")
print(f"Ожидаемое значение: {expected_diag:.2f}")
print(f"Расчет верен: {abs(calculated_diag - expected_diag) < 0.01}")

# Проверяем объем параллелепипеда 3x4x5: 3 * 4 * 5 = 60
calculated_volume = box1.calculate_volume()
expected_volume = 3 * 4 * 5
print(f"Объем параллелепипеда: {calculated_volume}")
print(f"Ожидаемое значение: {expected_volume}")
print(f"Расчет верен: {calculated_volume == expected_volume}")

print("\n" + "=" * 60)
print("ВСЕ ПРОВЕРКИ УСПЕШНО ЗАВЕРШЕНЫ!")
print("=" * 60)
```
